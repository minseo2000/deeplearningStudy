# String 모듈 임포트 해서 사용해보기

````python
import string

str1 = string.ascii_uppercase

print('str1=',str1)

str2 = str1[1:] + str1[:1]
print('str2 =', str2)

n = str1.index('A')
print(str2[n])
````

````python
import string

str1 = string.ascii_lowercase

print('str1=',str1)

str2 = str1[1:] + str1[:1]
print('str2 =', str2)

n = str1.index('a')
print(str2[n])
````

````python
import string

src_str = string.ascii_uppercase
dst_str = src_str[3:] + src_str[:3]

def cipher(a):
    idx = src_str.index(a)
    return dst_str[idx]


src = input('문장을 입력하시오: ')
print('암호화된 문장: ', end='')

for ch in src:
    if ch in src_str:
        print(cipher(ch), end='')
    else:
        print(ch, end='')
print()
````

````python
t = '''
[ARTICLE] 2000820 BLACKPINK Jennie is regarded to have great 
effet on KT Mystic Red as is was chosen by 50% of those who
prebooked for the Samsung Galaxy Note 20(LG U+ Mystic Pink 30%
SKT Mystic Blue not disclosed)
'''
low_t = t.lower()

for word in low_t.split(' '):
    if word == 'kt' or word =='samsung' or word == 'lg' or word =='skt':
        t += '** '
    else:
        t+= word +' '
print(t)
````

# 느낌표 개수, 대문자 개수세기


````python
t = '''
It's Not THe Right Time To Conduct Exams.
MY DEMAND IN BOLD AND CAPITAL. NO EXAMS IN COVID!!!
'''
count = 0
for ch in t:
    if ch.isupper():
        count+=1
    
print('느낌표 개수; ', t.count('!'))
print('대문자 개수: ', count)
````

# 숫자로만 구성되는 비밀번호 만들기!


````python
import random

n_digits = int(input('몇자리의 비밀번호 만들기: '))
otp = ''

for i in range(n_digits):
    otp+=str(random.randrange(0, 10))
print(otp)
````

# 문자와 숫자로 구성되는 비밀번호 만들기!

````python
import string
import random

lower_upper_digit = string.ascii_letters+'0123456789'




n_digits = int(input('몇자리의 비밀번호 만들기: '))
otp = ''

for i in range(n_digits):
    otp+= lower_upper_digit[(random.randrange(len(lower_upper_digit)))]
print(otp)
````

# 워드 클라우드 사용해보기

````python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

text = '''
[ARTICLE] 2000820 BLACKPINK Jennie is regarded to have great 
effet on KT Mystic Red as is was chosen by 50% of those who
prebooked for the Samsung Galaxy Note 20(LG U+ Mystic Pink 30%
SKT Mystic Blue not disclosed)
'''

s_w = set(STOPWORDS)

wc1 = WordCloud(stopwords = s_w, width=800, height=800)
wc1.generate(text)

plt.figure(figsize=(10, 8))
plt.imshow(wc1)
plt.show()
````

# wikipedia 사용해보기

````python
import wikipedia
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wiki = wikipedia.page('Monty Python')
text = wiki.content

wordcloud = WordCloud(width=2000, height=1500).generate(text)
plt.figure(figsize=(40, 30))
plt.imshow(wordcloud)
plt.show()
````

# 정규식

- 특정한 규칙을 가지고 있는 ㅁㄴ자열들을 표현하는데 사용되는 규칙을 가진 언어

````python
import re

txt1 = "Life is too short, you need python."
txt2 = "The best moments of my life."

print(re.search('Life', txt1))
print(re.search('Life', txt2))

print(re.search('Life|life', txt1))
print(re.search('Life|life', txt2))

print(re.search('[Ll]ife', txt1))
print(re.search('[Ll]ife', txt2))
````

# 문자열이 Life of life 로 시작되는가?  :  ^
# 문자열이 Life or life 로 끝나는가? : $

````python
import re

txt1 = "Life is too short, you need python."
txt2 = "The best moments of my life."
txt3 = "My life My Choice"

print(re.search('^Life|^life', txt1))
print(re.search('^Life|^life', txt2))
print(re.search('^Life|^life', txt3))

print(re.search('Life$|life$', txt1))
print(re.search('Life$|life$', txt1))
print(re.search('Life$|life$', txt1))
````

# 이메일 주소만 추리고 싶을 때!

````python
import re

txt = 'abc@facebook.com와 bbc@google.com에서 이메일이 도착하였습니다.'

print('추출된 이메일: ', re.findall('\S+@[a-z.]+', txt))
````

````python
import re

txt = 'abc@facebook.com와 bbc@google.com에서 이메일이 도착하였습니다.'

output = re.findall('\S+@[a-z.]+', txt)

for i in output:
    print(i.split('@'))
````

# 패스워드 검사 프로그램 만들기

````python
import re

while True:
    password = input('패스워드: ')
    if len(password)  < 8:
        not re.search('[A-Z]', password) or not re.search('[0-9]', password) or not re.search('[_@$!]', password):
            print('유효하지 않은 패스워드')
        else:
            print('유효한 패스워드')
            break
````
